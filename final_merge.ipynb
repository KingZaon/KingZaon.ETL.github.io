# Here We Are Merging DataFrame



```python
KickStarter = pd.DataFrame({'Total Kickstarter Industries': ['14'],
                            'Total Kaggle Industries':['26'],
                           'Total Funding': ['4,480,000,000'],
                           'Total Investors': ['16,663'],
                           'Success/Acquired': ['37.20']})
Kaggle = pd .DataFrame({'Total Kickstarter Industries': ['14'],
                        'Total Kaggle Industries': ['26'],
                       'Total Funding': ['454,045,620,000'],
                       'Total Investors': ['58,329'],
                       'Success/Acquired': ['336,927,339,047']})

both = pd.merge(KickStarter, Kaggle, on=['Total Kickstarter Industries', 'Total Kaggle Industries'])
```

# Combining Selected Data 



```python
final_merge = pd.merge(KickStarter, Kaggle, how='outer')
final_merge
```

```python
final_merge.to_sql('final_merge', con=engine)
```
